```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# iris dataset having multiple classisfication
df= sns.load_dataset('iris')
```

```python
df
```

```python
df['species'].unique()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop('species', axis=1)
```

```python
y=df['species']
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model_knn=KNeighborsClassifier(n_neighbors=13)
```

```python
model_knn.fit(X_train, y_train)
```

```python
model_knn.score(X_test,y_test)
```

```python
from sklearn.svm import SVC
```

```python
model_svm=SVC(gamma='auto')
```

```python
model_svm.fit(X_train, y_train)
```

```python
model_svm.score(X_test,y_test)
```

```python
#now lets use the grid search CV,  
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
# cross validation dont need train test split
classifier = GridSearchCV((model_svm), {
  'C' : [1,20,10,30],
    'kernel': ['rbf','linear'],
    
},cv=5,return_train_score=False)
```

```python
classifier.fit(X,y)
```

```python
classifier.cv_results_
```

```python
results=pd.DataFrame(classifier.cv_results_)
```

```python
results
```

```python
results[['param_C','param_kernel','mean_test_score']]
```

```python
classifier2 = GridSearchCV((model_knn), {
  'n_neighbors' : [1,20,10,30],
    'weights': ['uniform','distance'],
    
},cv=5,return_train_score=False)
```

```python
classifier2.fit(X,y)
```

```python
results=pd.DataFrame(classifier2.cv_results_)
```

```python
results
```

```python
results[['param_n_neighbors','param_weights','mean_test_score']]
```

```python
#Random Search CV
from sklearn.model_selection import RandomizedSearchCV
```

```python
classifier =RandomizedSearchCV((model_svm), {
  'C' : [1,20,10,30],
    'kernel': ['rbf','linear'],
    
},n_iter=4,cv=5,return_train_score=False)
```

```python
classifier.fit(X,y)
```

```python
results=pd.DataFrame(classifier.cv_results_)
```
